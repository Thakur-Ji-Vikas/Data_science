# Healthcare data

```python
import pandas as pd
data=pd.read_csv("Healthcare_data.csv")
data
```

```python
mean_age=data['Age'].mean()
median_age=data['Age'].median()
std_dev_age=data['Age'].std()
```

```python
print("Mean Age :",mean_age)
print("Median Age :",median_age)
print("Standard Deviation Age :",std_dev_age)
```

```python
# Calculate correlation between Age and Cholestrol
correlation_age_cholestrol=data['Age'].corr(data['Cholestrol'])

print("Correlation between Age and Cholestrol :",correlation_age_cholestrol)
```
