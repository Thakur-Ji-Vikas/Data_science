 # PCA (Principal Component Analysis)

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
from sklearn.decomposition import PCA
np.random.seed(0)
x=np.random.randn(100,2)
x

#initialize PCA with desired number of components
pca = PCA(n_components=1)
transformed_x = pca.fit_transform(x)
inv_transform = pca.inverse_transform(transformed_x)
inv_transform
```

```python
plt.scatter(inv_transform[:,0],inv_transform[:,1])
plt.show
```

# K-Means

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
# generating some random data points
np.random.seed(0)
x=np.random.rand(100,2)

# print(data)
# number of clusters

k=4


#kmeans model
kmeans = KMeans(n_clusters=k)

# fitting the k model
kmeans.fit(x)    

# get the cluster center and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

plt.scatter(x[:,0],x[:,1],c=labels,cmap='viridis',s=100)
plt.scatter(centers[:,0],centers[:,1],c='black',marker='*',s=100,label="Cluster centers")
plt.title('KMeans Clustering')
plt.legend(loc=1)
plt.show()
```
